```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
from bokeh.models import LinearAxis, Range1d, HoverTool, ColumnDataSource
```

```python
output_notebook()
```

```python
reactor = 3

source = ColumnDataSource(data=dict(
    x = pd.to_datetime(df.index, dayfirst=True),
    do = df.iloc[:,(reactor - 1)].values,
    ph = df.iloc[:,(reactor + 7)].values,
    acid = df.iloc[:,(14 + reactor * 2)].values,
    base = df.iloc[:,(15 + reactor * 2)].values
))
```

```python
p = figure(width=500, height=500, x_axis_type='datetime', tools="pan,reset,save,wheel_zoom")
p.extra_y_ranges={}
p.extra_y_ranges['do'] = Range1d(start = 0, end = 100)
p.extra_y_ranges['pH'] = Range1d(start = 0, end = 14)
p.add_layout(LinearAxis(y_range_name='pH', axis_label='pH'), 'right')
p.line('x', 'do', source=source, legend_label='do', line_width=2, color="blue", alpha=0.5)
p.line('x', 'ph', source=source, legend_label='pH', line_width=2, color="red", alpha=0.5, y_range_name='pH')
p.add_tools(HoverTool(tooltips=[('date', '@x{%F %T}'),('do','@do'),('pH', '@ph')],
          formatters={'@x': 'datetime'}))
show(p)
```

```python
p2 = figure(width=500, height=500, x_axis_type='datetime', y_range = (source.data['acid'].min(), source.data['acid'].max()))
base = source.data['base']
p2.extra_y_ranges={}
p2.extra_y_ranges['base'] = Range1d(start = base.min()-50, end = base.max()+50)
p2.add_layout(LinearAxis(y_range_name='base', axis_label='base'), 'right')
p2.line('x', 'acid', source=source, legend_label='acid', line_width=2, color="blue", alpha=0.5)
p2.line('x', 'base', source=source, legend_label='base', line_width=2, color="red", alpha=0.5, y_range_name='base')
p2.add_tools(HoverTool(tooltips=[('date', '@x{%F %T}'),('Acid','@acid'),('Base', '@base')],
          formatters={'@x': 'datetime'}))
show(p2)
```
